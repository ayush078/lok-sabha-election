```python
from bs4 import BeautifulSoup
import requests

# URLs to scrape
url_partywise = "https://results.eci.gov.in/AcResultGenJune2024/partywiseresult-S01.htm"
url_candidateswise = "https://results.eci.gov.in/AcResultGenJune2024/candidateswise-S01160.htm"

# Function to send request and parse content


def fetch_and_parse(url):
    response = requests.get(url)
    if response.status_code == 200:
        return BeautifulSoup(response.text, 'html.parser')
    else:
        print(f"Failed to retrieve data from {url}")
        return None


def extract_data(soup, class_name):
    extracted_data = []
    if soup:
        data_elements = soup.find_all('div', class_=class_name)
        for element in data_elements:
            extracted_data.append(element.text.strip())
    return extracted_data

# Function to extract party-wise data


def extract_partywise_data(soup):
    party_data = []
    if soup:
        rows = soup.find_all('tr')
        for row in rows:
            cols = row.find_all('td')
            if len(cols) >= 3:  # assuming at least 3 columns are present
                party_name = cols[0].text.strip()
                seats_won = cols[1].text.strip()
                party_data.append((party_name, seats_won))
    return party_data


# Fetch and parse content from the party-wise URL
soup_partywise = fetch_and_parse(url_partywise)
soup_candidateswise = fetch_and_parse(url_candidateswise)

# Extract data using the correct function
extracted_data_partywise = extract_partywise_data(soup_partywise)
extracted_data_candidateswise = extract_data(
    soup_candidateswise, 'col-md-4 col-12')


print("Party-wise Results:")
for party, seats in extracted_data_partywise:
    print(f"{party}: {seats}")
print("\nCandidates-wise Results:")
if len(extracted_data_candidateswise) == 0:
    print("YOLO")
for data in extracted_data_candidateswise:

    lines = [line for line in data.split('\n') if line.strip()]

    status = lines[0]  
    margin = lines[1] 
    name = lines[2] if len(lines) > 2 else ""
    party = lines[3] if len(lines) > 3 else "NA"

    # Printing the organized data
    print(f"status -> {status}")
    print(f"margin -> {margin}\n")
    print(f"name -> {name}")
    print(f"party -> {party}\n")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# URLs to scrape
url = "https://results.eci.gov.in/AcResultGen2ndJune2024/statewiseS021.htm"

# Function to send request and parse content
def fetch_and_parse(url):
    response = requests.get(url)
    if response.status_code == 200:
        return BeautifulSoup(response.text, 'html.parser')
    else:
        print(f"Failed to retrieve data from {url}")
        return None

# Function to extract data
def extract_data(soup, class_name):
    data = []
    if soup:
        table = soup.find('table', class_=class_name)
        rows = table.find_all('tr')[2:]  # Skipping the header rows
        for row in rows:
            cols = row.find_all('td')
            if len(cols) >= 9:  # Ensure there are enough columns
                constituency = cols[0].text.strip()
                const_no = cols[1].text.strip()
                leading_candidate = cols[2].text.strip()
                leading_party = cols[3].text.strip()
                trailing_candidate = cols[4].text.strip()
                trailing_party = cols[5].text.strip()
                margin_text = cols[6].text.strip().replace(',', '')
                
                # Handle the case where margin is not an integer
                try:
                    margin = int(margin_text)
                except ValueError:
                    margin = None
                
                round_no = cols[7].text.strip()
                status = cols[8].text.strip()
                data.append({
                    "Constituency": constituency,
                    "Constituency No.": const_no,
                    "Leading Candidate": leading_candidate,
                    "Leading Party": leading_party,
                    "Trailing Candidate": trailing_candidate,
                    "Trailing Party": trailing_party,
                    "Margin": margin,
                    "Round No.": round_no,
                    "Status": status
                })
    return data

# Fetch and parse the results
fetch_results = fetch_and_parse(url)
data = extract_data(fetch_results, "table table-striped table-bordered")

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)
print(df)
# # Remove rows with None margin
# df = df.dropna(subset=['Margin'])

# # Display the DataFrame
# print(df)

# # Create a bar graph of the margin of victory for each constituency
# plt.figure(figsize=(15, 10))
# plt.bar(df['Constituency'], df['Margin'], color='skyblue')
# plt.xlabel('Constituency')
# plt.ylabel('Margin of Victory')
# plt.title('Margin of Victory for Each Constituency')
# plt.xticks(rotation=90)
# plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# URLs to scrape
url = "https://results.eci.gov.in/AcResultGen2ndJune2024/statewiseS021.htm"

# Function to send request and parse content


def fetch_and_parse(url):
    response = requests.get(url)
    if response.status_code == 200:
        return BeautifulSoup(response.text, 'html.parser')
    else:
        print(f"Failed to retrieve data from {url}")
        return None

# Function to extract data


def extract_data(soup, class_name):
    data = []
    if soup:
        table = soup.find('table', class_=class_name)
        rows = table.find_all('tr')[2:]  # Skipping the header rows
        for row in rows:
            cols = row.find_all('td')
            if len(cols) >= 9:  # Ensure there are enough columns
                constituency = cols[0].text.strip()
                const_no = cols[1].text.strip()
                leading_candidate = cols[2].text.strip()
                leading_party = cols[3].text.strip()
                trailing_candidate = cols[4].text.strip()
                trailing_party = cols[5].text.strip()
                margin_text = cols[6].text.strip().replace(',', '')

                # Handle the case where margin is not an integer
                try:
                    margin = int(margin_text)
                except ValueError:
                    margin = None

                round_no = cols[7].text.strip()
                status = cols[8].text.strip()
                data.append({
                    "Constituency": constituency,
                    "Constituency No.": const_no,
                    "Leading Candidate": leading_candidate,
                    "Leading Party": leading_party,
                    "Trailing Candidate": trailing_candidate,
                    "Trailing Party": trailing_party,
                    "Margin": margin,
                    "Round No.": round_no,
                    "Status": status
                })
    return data


# Fetch and parse the results
fetch_results = fetch_and_parse(url)
data = extract_data(fetch_results, "table table-striped table-bordered")

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)
# print(df)



# Create a bar graph of the margin of victory for each constituency
plt.figure(figsize=(15, 10))
plt.bar(df['Constituency'], range(len(df['Constituency'])), color='skyblue')
plt.xlabel('Constituency')
plt.ylabel('Margin of Victory')
plt.title('Margin of Victory for Each Constituency')
plt.xticks(rotation=90)
plt.show()
```
